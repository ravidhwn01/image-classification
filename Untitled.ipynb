```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow  as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = 'Fruits_Vegetables/train'
data_test_path = 'Fruits_Vegetables/train'
data_val_path = 'Fruits_Vegetables/train'
```

```python
# set our images in a same dimensions(because all images are in diff. dimensions) convert into fixed width and height
img_width = 180
img_height = 180

```

```python
# bring image data into form of arrays
data_train  = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle = True,
    image_size=(img_width,img_height),
    batch_size=32,
    validation_split=False
)
```

```python
 data_category = data_train.class_names
```

```python
data_category
```

```python
# bring image data into form of arrays
data_validataion  = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    shuffle = False,
    image_size=(img_width,img_height),
    batch_size=32,
    validation_split=False
)
```

```python
# bring image data into form of arrays
data_test  = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle = False,
    image_size=(img_width,img_height),
    batch_size=32,
    validation_split=False
)
```

```python
# lets print some images along with labels from training dataset
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_category[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
#create model
model = Sequential([
    layers.Rescaling(1./255, input_shape=(img_width,img_height,3)),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_category))
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.summary()
```

```python
epochs = 25
history = model.fit(
    data_train,
    validation_data = data_validataion,
    epochs = epochs,
    
)
```

```python
# chart between training accuracy and validation dataset accuracy
epochs_range = range(epochs)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range, history.history['accuracy'], label='Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title(' Accuracy')


plt.subplot(1,2,2)
plt.plot(epochs_range, history.history['loss'], label='Training loss')
plt.plot(epochs_range, history.history['val_loss'], label='Validation loss')
plt.legend(loc='lower right')
plt.title('Loss')
```
